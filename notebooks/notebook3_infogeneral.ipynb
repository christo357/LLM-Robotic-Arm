## EzPickle
What EzPickle does

When you call:

EzPickle.__init__(self, reward_type=reward_type, **kwargs)

it records the constructor arguments so that, if the env is pickled, Gym can later do something like:

“reconstruct the env by calling the same class with the same args”

This shows up in a few situations:
	•	Multiprocessing with SubprocVecEnv (envs must be pickleable on many platforms)
	•	Saving and loading “vectorized env constructors”
	•	Some wrappers and monitor utilities that rely on env pickling
	•	RL libraries that spawn worker processes
